# Visualisaatio lineaarisesti

```python
from matplotlib.colors import ListedColormap
import matplotlib.pyplot as plt

def plot_decision_regions(X, y, classifier, test_idx=None,
                          resolution=0.02):
    # setup marker generator and color map
    markers = ('s', 'x', 'o', '^', 'v')
    colors = ('red', 'blue', 'lightgreen', 'gray', 'cyan')
    cmap = ListedColormap(colors[:len(np.unique(y))])
    
    # plot the decision surface
    x1_min, x1_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    x2_min, x2_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx1, xx2 = np.meshgrid(np.arange(x1_min, x1_max, resolution),
                           np.arange(x2_min, x2_max, resolution))
    
    Z = classifier.predict(np.array([xx1.ravel(), xx2.ravel()]).T)
    Z = Z.reshape(xx1.shape)
    plt.contourf(xx1, xx2, Z, alpha=0.3, cmap=cmap)
    plt.xlim(xx1.min(), xx1.max())
    plt.ylim(xx2.min(), xx2.max())

    for idx, cl in enumerate(np.unique(y)):
        plt.scatter(x=X[y == cl, 0], y=X[y == cl, 1],
                    alpha=0.8, c=colors[idx],
                    marker=markers[idx], label=cl,
                    edgecolor='black')

        # highlight test examples with circles
        if test_idx:
            # plot all examples
            X_test, y_test = X[test_idx, :], y[test_idx]

            plt.scatter(X_test[:, 0], X_test[:, 1],
                        c='', edgecolor='black', alpha=1.0,
                        linewidth=1, marker='o',
                        s=100, label='test set')

#Specify the indices
X_combined_std = np.vstack((X_train_std, X_test_std))
y_combined = np.hstack((y_train, y_test))
plot_decision_regions(X=X_combined_std,
                      y=y_combined,
                      classifier=ppn,
                      test_idx=range(105, 150))

plt.xlabel('petal length [standardized]')
plt.ylabel('petal width [standardized]')
plt.legend(loc='upper left')
plt.tight_layout()
plt.show()
```

# Luokkien todennäköisyys logistisella regressiolla

Adaline vs. Logistinen regressio



Sigmoidifunktion tulos tulkitaan sitten luokkaan 1 kuuluvan tietyn esimerkin todennäköisyydeksi **𝜙(𝑧) = 𝑃(𝑦 = 1 | 𝒙; 𝒘)**, kun otetaan huomioon sen ominaisuudet, **x**, joka on parametroitu painoilla **w**.

**Esimerkiksi**, jos laskemme tietylle kukalle **𝜙(𝑧) = 0.8**, se tarkoittaa, että mahdollisuus tämän esimerkin olevan *Irisversicolor-kukka*, on **80 prosenttia.**

Siksi todennäköisyys, että tämä kukka on *Iris-setosa-kukka*, voidaan laskea seuraavasti: **𝑃(𝑦 = 0 | 𝒙; 𝒘) = 1 - 𝑃(𝑦 = 1 | 𝒙; 𝒘)** = 0.2 tai **20 prosenttia.**

Ennustettu todennäköisyys voidaan sitten yksinkertaisesti muuntaa binääriseksi tulokseksi kynnysfunktion kautta.

## Tarkempi selitys

Perceptronin yksi ongelmista on ettei se koskaa lähenny jos luokkie ei voi erottaa lineaarisesti toisistaan, koska väärinluokitteluja on aina vähintään yksi jokaisessa epochissa joten painoja päivitetään jatkuvasti.

Logistinen regressio tarvitsee todennäköisyyden (odds) johon lisätään logit-funktio:



p on positiivisen tapahtuman todennäiköisyyttä (jotain minkä voi ennustaa)

Nyt voidaan ilmaista lineaarinen suhde ominaisuusarvojen ja log-kertoimien välillä:



𝑝(𝑦 = 1|𝒙) on ehdollinen todennäköisyys missä tietty esimerkki kuuluu luokkaan 1 sen ominaisuuksien x vuoksi.

Ennustetaan todennäköisyys, missä tietty esimerkki kuuluu tiettyyn luokkaan, mikä on *logit-funktion käänteinen muoto* eli **logistiseksi sigmoidi funktio** (joskus yksinkertaisesti lyhennetään **sigmoidiksi funktioksi** sen tyypillisen S-muodon vuoksi):



z on nettosyöttö, painojen lineaarinen yhdistelmä ja syöttö (harjoitusesimerkkeihin liittyvät ominaisuudet):

# Plotataan tämä S-muotoinen (sigmoidinen) käyrä

```python
import matplotlib.pyplot as plt
import numpy as np
def sigmoid(z):
    return 1.0 / (1.0 + np.exp(-z))
z = np.arange(-7, 7, 0.1)
phi_z = sigmoid(z)
plt.plot(z, phi_z)
plt.axvline(0.0, color='k')
plt.ylim(-0.1, 1.1)
plt.xlabel('z')
plt.ylabel('$\phi (z)$')
# y axis ticks and gridline
plt.yticks([0.0, 0.5, 1.0])
ax = plt.gca()
ax.yaxis.grid(True)
plt.tight_layout()
plt.show()
```

Kuten näkyy; sigmoidifunktio ottaa reaalilukuarvot syötteeksi ja muuntaa ne arvoiksi alueella [0, 1] leikkauksella 𝜙(𝑧) = 0,5.

# Visualisaatio yhden harjoitusesimerkin eri arvojen 𝜙(𝑧) luokittelu kustannuksista

Log-todennäköisyys kustannusfunktiona **J**, joka voidaan minimoida käyttämällä gradientteja:



```python
def cost_1(z):
    return - np.log(sigmoid(z))
def cost_0(z):
    return - np.log(1 - sigmoid(z))

z = np.arange(-10, 10, 0.1)
phi_z = sigmoid(z)
c1 = [cost_1(x) for x in z]
plt.plot(phi_z, c1, label='J(w) if y=1')
c0 = [cost_0(x) for x in z]
plt.plot(phi_z, c0, linestyle='--', label='J(w) if y=0')
plt.ylim(0.0, 5.1)
plt.xlim([0, 1])
plt.xlabel('$\phi$(z)')
plt.ylabel('J(w)')
plt.legend(loc='best')
plt.tight_layout()
plt.show()
```

# Adalinen muuntaminen logistisen regression algoritmiksi

```python
class LogisticRegressionGD(object):

    def __init__(self, eta=0.05, n_iter=100, random_state=1):
        self.eta = eta
        self.n_iter = n_iter
        self.random_state = random_state

    def fit(self, X, y):
        rgen = np.random.RandomState(self.random_state)
        self.w_ = rgen.normal(loc=0.0, scale=0.01,
                              size=1 + X.shape[1])
        self.cost_ = []

        for i in range(self.n_iter):
            net_input = self.net_input(X)
            output = self.activation(net_input)
            errors = (y - output)
            self.w_[1:] += self.eta * X.T.dot(errors)
            self.w_[0] += self.eta * errors.sum()

            # note that we compute the logistic 'cost' now
            # instead of the sum of squared errors cost
            cost = (-y.dot(np.log(output)) -
                    ((1 - y).dot(np.log(1 - output))))
            self.cost_.append(cost)
        return self

    def net_input(self, X):
        '''Calculate net input'''
        return np.dot(X, self.w_[1:]) + self.w_[0]

    def activation(self, z):
        """Compute logistic sigmoid activation"""
        return 1. / (1. + np.exp(-np.clip(z, -250, 250)))

    def predict(self, X):
        """Return class label after unit step"""
        return np.where(self.net_input(X) >= 0.0, 1, 0)
        '''
        Equivalent to:
        return np.where(self.activation(self.net_input(X)) >= 0.5, 1, 0)
        '''
```

```python
X_train_01_subset = X_train[(y_train == 0) | (y_train == 1)]
y_train_01_subset = y_train[(y_train == 0) | (y_train == 1)]

lrgd = LogisticRegressionGD(eta=0.05,
                            n_iter=1000,
                            random_state=1)

lrgd.fit(X_train_01_subset, y_train_01_subset)
plot_decision_regions(X=X_train_01_subset,
                      y=y_train_01_subset,
                      classifier=lrgd)

plt.xlabel('petal length [standardized]')
plt.ylabel('petal width [standardized]')
plt.legend(loc='upper left')
plt.tight_layout()
plt.show()
```

# Ylisovittamisesta



#### Yksi tapa löytää hyvä bias-variant-tradeoff (kuvaa koneoppimismallin suorituskykyä) on virittää mallin monimutkaisuus säätöjen (regulation) avulla.

Säädön taustalla on lisätietojen (ennakkoluulojen = bias) tuominen äärimmäisten parametriarvojen (paino) rankaisemiseksi. 

Yleisin säädön muoto on niin kutsuttu **L2-säätö** (joskus kutsutaan myös L2-kutistumiseksi tai painon heikkenemiseksi):



𝜆 on ns. säätöparametri.

Logistisen regression kustannusfunktio voidaan säännellä lisäämällä yksinkertainen säätötermi, joka kutistuu painot mallikoulutuksen aikana:



Säätöparametrin 𝜆 avulla voimme sitten hallita kuinka hyvin sovitamme harjoitustiedot pitäen painot pieninä. Lisäämällä 𝜆:n arvoa lisätään säädön vahvuutta.

# Visualisaatio L2-säätöreitistä kahdelle painokertoimelle

```python
weights, params = [], []
for c in np.arange(-5, 5):
    #sovitetaan 10 logistista regressiomallia, joilla oli erilaiset arvot *käänteis-säätöparametrille* **C** (normaali säätöparametri oli 𝜆).
    #fits 10 logistic regression models with different values for the inverse-regulation-parameter C (normal legalization-parameter was 𝜆).
    lr = LogisticRegression(C=10.**c, random_state=1,
                            solver='lbfgs', multi_class='ovr')
    lr.fit(X_train_std, y_train)
    weights.append(lr.coef_[1])
    params.append(10.**c)
    
weights = np.array(weights)
```

```python
plt.plot(params, weights[:, 0],
         label='petal length')

plt.plot(params, weights[:, 1], linestyle='--',
         label='petal width')

plt.ylabel('weight coefficient')
plt.xlabel('C')
plt.legend(loc='upper left')
plt.xscale('log')
plt.show()
```

### Painokertoimet kutistuvat, jos pienennämme parametria *C*, toisin sanoen, jos lisäämme säädön vahvuutta.

Kuvassa on tosin vain luokan 1 (tässä tietojoukon toisen luokan: Iris-versicolor) painokertoimet verrattuna kaikkiin luokittelijoihin.
